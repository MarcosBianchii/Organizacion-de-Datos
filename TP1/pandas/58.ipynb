# Observe una muestra aleatoria de los comentarios de las acciones realizadas por usuarios o ips antes de ser bloqueados. Observe otra muestra de comentarios de acciones de todos. (⭐⭐)

```python
import pandas as pd
logs = pd.read_csv('../data/logs.csv', low_memory=False, usecols=['timestamp','contributor_username','contributor_id','contributor_ip','comment','action','title'])
logs
```

```python
logs.rename(columns={'contributor_username':'user','contributor_id':'id','contributor_ip':'ip','comment':'comentario','action':'accion','title':'titulo'}, inplace=True)
logs_copia = logs.copy()
```

```python
# Quito los campos con id nulo porque no voy a poder identificar a su usuario.
logs = logs[~logs.id.isnull()]
# Quito los campos NaN
logs = logs.fillna('')
logs
```

```python
logs.id = logs.id.astype('Int32')
```

```python
# Junto los campos user e ip para traducir los nombre/ip a ids.
logs['identificador'] = logs['user'] + logs['ip']
logs = logs.drop(['user','ip'], axis=1)
logs
```

```python
# Me quedo con los campos de bloqueos.
bloqueos = logs[logs.accion == 'block'].drop(columns=['accion','comentario','identificador','id'], axis=1)
```

```python
# Me quedo con las fechas de los bloqueos y a su respectivo bloqueado.
bloqueos.titulo = bloqueos.titulo.map(lambda x: x.replace('Usuario:',''))
bloqueos.rename(columns={'titulo':'identificador'}, inplace=True)
bloqueos
```

```python
# Uso este DF para traducir el identificador de bloqueados a ids.
usuarios = logs[['identificador','id']]
usuarios
```

```python
# Consigo un DF con los ids de los bloqueados y sus fechas de bloqueo.
bloqueados = pd.merge(bloqueos, usuarios, on='identificador')
bloqueados.drop(columns=['identificador'], axis=1, inplace=True)
bloqueados
```

```python
# Agrupo y consigo el primer bloqueo de cada usuario.
bloqueados = bloqueados.groupby('id').agg({'timestamp':'min'}).reset_index()
bloqueados
```

```python
# Junto los DFs para obtener los comentarios de los bloqueados junto con sus fechas.
bloqueados = pd.merge(bloqueados, logs[['id','timestamp','comentario']], on='id')
bloqueados.rename(columns={'timestamp_x':'fecha_bloqueo','timestamp_y':'fecha_comentario'}, inplace=True)
```

```python
# Saco los comentarios vacios (eran NaN) para que al mostrarlos no sean todos comentarios vacios.
bloqueados = bloqueados[bloqueados.comentario != '']
bloqueados
```

```python
# Filtro los comentarios que fueron posteriores a su primer bloqueo.
bloqueados = bloqueados[bloqueados.fecha_bloqueo > bloqueados.fecha_comentario]
bloqueados
```

## Muestro un sample de comentarios de usuarios antes de ser bloqueados.

```python
bloqueados.comentario.sample(20)
```

## Muestro un sample de comentarios de todos los usuarios.

```python
# Quito los comentarios NaN para mostrar comentarios con contenido.
logs_copia[~logs_copia.comentario.isnull()].comentario.sample(20)
```
