```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cierre = pd.read_excel('data/datasetRofex4.xlsx').Cierre
cierre
```

```python
plt.figure(dpi=125)
sns.histplot(cierre, bins=40, color='#287233', edgecolor='#287233')
plt.xlabel('Precio de ajuste del día')
plt.ylabel('Cantidad')
plt.title('Cantidad de veces que el precio de la tonelada\n de soja cerro en un precio especifico')
plt.show()
```
